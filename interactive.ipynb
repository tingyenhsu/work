```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def f(x):
    return x
```

```python
interact(f, x=widgets.IntSlider(min=-10, max=30, step=1, value=10));
```

```python
interact(f, x=True);
```

```python
interact(f, x='Hi there!');
```

```python
@interact(x=True, y=1.0)
def g(x, y):
    return (x, y)
```
